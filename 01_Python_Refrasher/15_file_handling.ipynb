File handling allows Python programs to read, write, and manipulate files stored on disk. Python provides built-in functions for working with files.

## Summary of File Operations

| Operation | Description | Example |
|-----------|-------------|---------|
| Open file | Opens a file in the specified mode. | `file = open("example.txt", "r")` |
| Read file | Reads the entire contents of the file. | `file.read()` |
| Read line | Reads the next line from the file. | `file.readline()` |
| Read lines | Reads all lines into a list. | `file.readlines()` |
| Write file | Writes content to the file (overwrites existing content in write mode). | `file.write("Hello")` |
| Append file | Appends content to the end of the file. | `file.write("\nMore text")` |
| Check file existence | Checks whether a file exists before opening or deleting it. | `os.path.exists("file.txt")` |
| Delete file | Deletes a file from the filesystem. | `os.remove("file.txt")` |

```python
with open("text.txt", "r") as f:
    print(f.read())
```
