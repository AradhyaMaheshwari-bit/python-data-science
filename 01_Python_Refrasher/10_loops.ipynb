## For Loop
Used to iterate over sequences like lists, tuples, and strings.

```python
i = [3,5,6,67]
```

```python
for item in i:
    print(item)
```

Using `range()`

```python
for i in range(5,11):
    print(i)
```

## While Loop
Runs as long as a condition is True.

```python
i = 0
while i < 12:
    print(i)
    i += 1  # i = i+1
```

## Loop Control Statements
`break` → Exits the loop.

`continue` → Skips to the next iteration.

`pass` → Does nothing (used as a placeholder).

```python
i = ["apples", "bananas", "tomatoes"]
for item in i:
    if (item == "bananas"):
        break  # stops loop at  bananas
    print(item)
print("Hello")
```

```python
i = ["apples", "bananas", "tomatoes"]
for item in i:
    pass     # don't do anything (placeholder)
print("Hello")
```

```python
i = ["apples", "bananas", "tomatoes"]
for item in i:
    if (item == "bananas"):
        continue   # skip this iteration of the loop
    print(item)
print("Hello")
```
